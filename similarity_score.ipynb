```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
doc1 = nlp(u'Hello hi there!')
doc2 = nlp(u'Hello hi there!')
doc3 = nlp(u'Hey whatsup?')

print(doc1.similarity(doc2)) # 0.999999954642
print(doc2.similarity(doc3)) # 0.699032527716
print(doc1.similarity(doc3)) # 0.699032527716
```

```python
df = pd.read_csv('./dataset/meta.csv')
df.head(10)
```

```python
import numpy as np
df['text'] = df['title'] +df['date']+df['description']+df['content']
j=0
for title in df['text']:
    t1 = nlp(title)
    
    idx = None
    idlist = list()
    
    maxi = 0.95
    i = 0
    for another_title in df['text']:
        t2 = nlp(another_title)
        if maxi < t1.similarity(t2) and t1.similarity(t2) != 1.0:
            maxi = t1.similarity(t2)
            idx = i
        i+=1
    allowed_dev = 0.99*maxi if maxi is not None else -np.inf
    i = 0
    for another_title in df['text']:
        t2 = nlp(another_title)
        if allowed_dev < t1.similarity(t2) and t1.similarity(t2) != 1.0:
#             maxi = t1.similarity(t2)
            idlist += [i]
        i+=1
    
    print("actual : ",df['title'][j])
    if idx is not None:
        print("closest : ",df['title'][idx])
    for temp in idlist:
        if temp is not idx:
            print("Others : ",df['title'][temp])
    print("\n")
    j+=1
```
