# Project Spectacle

## Approach

#### Building Scraping algo:
1. Manually build two datasets:
    - url_viral.csv
    - url_scrape.csv
1. Use meta_info to generate the datasets with all the required infor and meta data. Output is:
    - viral.csv
    - scrape.csv
1. Clustering
1. Tuning

#### Usage:
1. Once the model is tuned generate a url_viral.csv with all the viral news
1. Scrape the url_viral.csv to generate viral.csv
1. Scrape google(top 2 pages) for each row in viral.csv for each date in range +-5 from viral articles given date.
1. Generate meta_info to scrape out relevant information for each link.
1. Use the similarity score and threshold to classify the articles.
1. Discard the outliers.
1. Use this as dataset for analysis.



## Clustering Options:
### NLP + KNN
- Use KNN, with K=1 to cluster the scrape.csv
- pseudocode:
    ```
    for each row in test dataset:
        generate similarity score with each row in training set
        if similarity score is in given threshold:
            set cluster number for one with closes match
        else:
            set cluster number to that of the outlier
    ```
- Generate success rate with regards to different distance threshold

#### Con:
1. All articles might not necessarily have best similarity score with one in viral set. They might be closer to similar article in scrape.csv
1. Similarity score needs to be weighted heavily on parameters like Date of publishment
