```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint as pp
from urllib.parse import urlparse, parse_qs
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import platform
path = None
if platform.system() is 'Windows':
    path = './drivers/geckodriver.exe'
else:
    path = './drivers/geckodriver'
options = Options()
options.headless = True
driver = webdriver.Firefox(options=options,executable_path=path)
```

```python
URL = 'https://www.google.com/search?q={query}&hl=en&gl=us&as_drrb=b&authuser=0&ie=UTF-8&tbm=nws&ei=c7rVXejdEMTf9QOtuaa4CQ&start={start}&sa=N'
query="Egypt"


def get_page_source(url):
    driver.get(url)
    return driver.page_source

def run(**params):
    response = get_page_source(URL.format(**params))
    return response

def scrape(soup,urls):
    temp = set()
    body = soup.findAll('g-card')
    for test in body:
        urls.add(test.find('a').attrs['href'])

def search(urls):
    for i in range(0,10):
        soup = BeautifulSoup(run(query=query,start=i*10),'html.parser')
        scrape(soup, urls)
```

```python
links = set()   
search(links)
for val in links:
    print(val)
```
